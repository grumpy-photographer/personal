```python
import pandas as pd
import geopandas as gps
```

```python
data = pd.read_csv("https://data.nasa.gov/api/views/gh4g-9sfh/rows.csv?accessType=DOWNLOAD")
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
dummies = pd.get_dummies(data['fall'])
dummies
```

```python
#data_clean = data.drop("fall", axis =1)
data_dum = data.join(dummies).drop("fall", axis =1)
```

```python
data_dum = data_dum.select_dtypes(exclude = "object")
data_dum = data_dum.drop("id", axis = 1)
```

```python
data_dum = data_dum.dropna()
```

```python
X = data_dum.drop("Fell", axis =1)
y = data_dum["Fell"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc = RandomForestRegressor().fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

```python
lr_pred = lr.predict(X)
```

```python
rfc.score(X_test, y_test)
```
