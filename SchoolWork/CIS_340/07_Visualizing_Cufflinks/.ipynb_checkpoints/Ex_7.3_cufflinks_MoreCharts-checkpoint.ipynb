```python
from IPython.display import display, HTML
import pandas as pd
import numpy as np
import math

# Using plotly + cufflinks in offline mode

import plotly.plotly as py
import plotly.graph_objs as go
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
cf.set_config_file(offline=True)
```

```python
df = pd.read_csv('Data\\Country_Data.csv')
df.head()
```

# Ex Chart 1: Population Map for the Americas:  2007
- Data Shape:  Tall (Not Pivoted!)
- Timeframe:  2007 
- Additional Filtering:  Continent: Americas
- Chart Type: Choropleth  
- Fill coloring based on:  Population

```python
df.head(2)

df['year'].unique()
```

```python
df_graph = df[(df['year']==2007) & (df['continent']=='Americas')]

# Sort by Highest population to lowest
df_graph.sort_values('population', ascending=False, inplace=True)
df_graph.head()
```

```python
# Choropleth Maps for Per Capita GDP
df_graph.iplot(
    kind='choropleth', 
    locations='country',  
    z ='population',
    text = 'country', 
    locationmode = 'country names', 
    theme='white',
    colorscale='greens', 
    title = "7.3 Ex Chart 1: Population Map for Americas: 2007",
    projection = dict(
            type = 'natural earth'
        ))
```

# Ex Chart 2: Scatter Plot of GDP per Capita vs Life Expectancy  
- Data Shape:  Tall (Not Pivoted!)
- Timeframe:  All the years in df
- Additional Filtering:  None
- Chart Type: Scatter Plot  
- x Axis:  GDP per Capita  
- y Axis: Life Expectancy

```python
df.iplot(kind='scatter', 
         x='gdpPerCapita',
         y='lifeExpectancy',
         mode='markers',
         symbol='circle-dot',
         colors=['orange','teal'],
         size=10,
         xTitle='Life Expectancy',
         yTitle='Per Capita GDP',
         title='7.3 Ex Chart 2: Scatter Plot of GDP per Capita vs Life Expectancy')
```

# Ex Chart 3: Bubble Chart GDP per Capita vs Life Expectancy- Population (2007)  
- Data Shape:  Tall (Not Pivoted!)
- Timeframe:  2007
- Additional Filtering:  None
- Chart Type: Bubble Chart  
- x Axis:  GDP per Capita  
- y Axis: Life Expectancy  
- Size of Bubbles:  Population

```python
df_graph2 = df[(df['year']==2007)]

df_graph2.sort_values('population', ascending=False, inplace=True)

df_graph2.iplot(kind='bubble',
                x='gdpPerCapita',
                y='lifeExpectancy',
                size='population', 
                categories='continent',
                text='continent',
                theme='solar',
                xTitle='GDP per Capita',
                yTitle='Life Expectancy',
                title='7.3 Ex Chart 3: Bubble Chart GPD per Capita vs Life Expectancy- Population (2007)')
```

# Extra Credit (10 points): Ratio Chart for Thailand vs Vietnam 
- This uses the pivoted df_wide_GDP dataframe

```python
df_wide_GDP = df.pivot(index='year', columns='country', values='gdpPerCapita')
```

```python
countries_to_plot = ['Thailand', 'Vietnam']

df_wide_GDP[countries_to_plot].iplot(kind='ratio', 
                                     yTitle='GDP per Capita',
                                     colors=['green','red'],
                                     title='GDP per Capita: Thailand vs Vietnam')
```
