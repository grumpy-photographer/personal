```python
#Imports
import pandas as pd
```

```python
#Load Land data
df_mlsf = pd.read_excel('./Data/County_MedianListingPricePerSqft_AllHomes.xlsx')

#Display table to ensure data loaded correctly
df_mlsf.head()
```

```python
#Filter data to NC
filter1 = df_mlsf['State'] == "NC"
df_mlsf_nc = df_mlsf[filter1]

#Check to ensure filter worked
df_mlsf_nc.head(5)
```

```python
#Change MunicipalCodeFIPS dtype to add leading 0's
df_mlsf_nc['MunicipalCodeFIPS'] = df_mlsf_nc['MunicipalCodeFIPS'].astype(str)
```

```python
#Add leading 0's and check to ensure they were added
df_mlsf_nc['MunicipalCodeFIPS'] = df_mlsf_nc['MunicipalCodeFIPS'].str.zfill(3)
df_mlsf_nc.head(5)
```

```python
#Save to csv file for export in Excel
df_mlsf_nc.to_csv('./Data/STG_ZLLW_County_MedianListingPricePerSqft_AllHomes.txt')
```
