# Dataframe to Start With

```python
# Data to work on
df_employees = pd.DataFrame(
    [
    [1702,'South','Robinett, David','david22@adventure-works.com', '(827) 525-0100', '06-05-2010', '$80,950', 'dsfkjdfl','dsfkjdfl'], 
    [1703,'North','Roberson, Rebecca','rebecca5 @adventure-works.com', '(829) 525-0101', '05-01-2015', '$70,950','dsfkjdfl','dsfkjdfl'],
    [1704,'North','Zillo, Dan','dan3@adventure-works.com', '(828) 555-0102', '03-01-2017', '$50,000','dsfkjdfl','dsfkjdfl'],
    [1700,'South','Craft, Reba','rebea@adventure-works.com', '(829) 525-0101', '05-01-2015', '$70,950','dsfkjdfl','dsfkjdfl'],
    [1698,'South','Craft, Maria','maria5@adventure-works.com', '(829) 525-0101', '05-01-2015', '$70,950','dsfkjdfl','dsfkjdfl'],
    [1699,'South','Adams, Jorge','adams9@adventure-works.com', '(827) 525-0100', '06-05-2010', '$80,950', 'dsfkjdfl','dsfkjdfl'],   
    [1696,'South','Forney, Dorothy','dorothy3@adventure-works.com', '(828) 555-0102', '03-01-2017', '$50,00','dsfkjdfl','dsfkjdfl'],
    [1701,'South','Forbes, Donna','donna@adventure-works.com', '(828) 555-0102', '03-01-2017', '$50,00','dsfkjdfl','dsfkjdfl']
    ], 
    columns=['ID', 'Region', 'employee_name','email_address', 'PhoneNumber', 'StartDate', 'AnnualSalary', 'Junk_1', 'Junk_2'])
  
df_employees
```

# Dropping Columns  
Drop the following columns: **Junk_1** and **Junk_2** 

```python
# drop columns
df_employees.drop(['Junk_1', 'Junk_2'], axis='columns', inplace=True)

# display results
df_employees.head()
```

# Renaming Columns  
Rename the following columns:  
1. **ID** to **BusinessEntityID**
2. **employee_name** to **EmployeeName**   
3. **email_address** to **EmailAddress** 

```python
# create new column library
new_columns = {'ID':'BusinessEntityID',
              'employee_name':'EmployeeName',
              'email_address':'EmailAddress'}

# rename columns
df_employees.rename(columns = new_columns, inplace=True)
df_employees.head()
```

# Reordering Columns  
Reorder the columns to the following order (left to right): 
1. Employee name  
2. Start date of employement  
3. Region  
3. Email address  
4. Phone number  
5. Current salary  
7. BusinessEntityID

```python
# Create new column order
new_cols = ['EmployeeName', 'StartDate', 'Region', 'EmailAddress', 'PhoneNumber', 'AnnualSalary', 'BusinessEntityID']
new_cols
```

```python
# reorder columns
df_employees = df_employees[new_cols]
df_employees.head()
```

# Split Columns  
- Split the **EmployeeName** column into two columns:  
  - **LastName**  
  - **FirstName**

```python
df_temp = df_employees['EmployeeName'].str.split(',', 1, expand=True)
df_temp.head()
```

```python
# Name columns for df_temp:  StateAbbrev and SubstateID
df_temp.columns = ['LastName', 'FirstName']

# display
df_temp.head()
```

```python
# Concatenate
df_employees2 = pd.concat([df_employees, df_temp], axis=1)

# Display top rows
df_employees2.head()
```

# Get Rid of Non-Numeric Junk  
- Get rid of the non-numeric junk in any column that contains dollar/currency values

```python
# Get rid of non-numeric junk
df_employees2['AnnualSalary'] = df_employees2['AnnualSalary'].str.replace('$', '')
df_employees2['AnnualSalary'] = df_employees2['AnnualSalary'].str.replace(',', '')
df_employees2['AnnualSalary'].head()
```

# Change Data Types  
1. Change the data type of any column that contains dollar/currency values to a float  
2. Change the data type of any column that contains Date values to a DateTime

```python
df_employees2.dtypes
```

```python
# Change AnnualSalary to float
df_employees2['AnnualSalary'] = pd.to_numeric(df_employees2['AnnualSalary']).astype(float)

df_employees2.dtypes
```

```python
# Change StartDate to date
df_employees2['StartDate'] = pd.to_datetime(df_employees2['StartDate'])

df_employees2.dtypes
```

# Subset Dataframe  
Based on the current dataframe, create a new dataframe that only has employees from the **South** Region

```python
filter = df_employees2['Region'] == 'South'

df_south = df_employees2[filter]

df_south.head()
```

# Sort  
Sort the subsetted dataframe by last name and first name, both from a-z

```python
df_south.sort_values(['LastName','FirstName'], inplace=False, ascending=True)
```

# Create New Column  
- In the subsetted dataframe, create a new column named **start_year** that is the year of the StartDate 

```python
df_south['start_year'] = df_south['StartDate'].dt.year
df_south.head()
```

# Do Calculations based on the Subsetted Dataframe
- Note:  You do NOT have to format your calculated numbers!

### Calculate and display the sum of the AnnualSalary column

```python
sum_salary = df_south['AnnualSalary'].sum()
print("The sum of annual salary is ", sum_salary)
```

### Calculate and display the sum of the AnnualSalary column for the year 2015

```python
filter = df_south['start_year'] == 2015
df_2015 = df_south[filter]
df_2015
```

```python
sum_salary_2015 = df_2015['AnnualSalary'].sum()
print("The sum of annual salary from 2015 is ", sum_salary_2015)
```

### Calculate the percentage of the total AnnualSalary that  is from 2015

```python
percentage = df_south['AnnualSalary']/df_2015['AnnualSalary']
print("The percentage of annual salary from 2015 is ", percentage)
```
