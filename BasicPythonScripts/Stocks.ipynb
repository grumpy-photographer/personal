```python
import matplotlib.pyplot as plt
import pandas as pd

import pandas_datareader.data as web
import datetime as dt
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
cf.set_config_file(offline=True)
```

```python
# Start for Stock series:  Jan 1, 2010
start = dt.datetime(2010, 1, 1)

# End of Stock series:  Dec 31, 2018
end = dt.datetime(2019, 8, 31)
```

```python
stocks_to_retrieve = ['IBM', 'AAPL', 'NFLX', 'ORCL', 'MSFT']
df = web.DataReader(stocks_to_retrieve, 'yahoo', start, end)
df.head()
```

```python
df_closing = df['Close']
df_closing.head()
```

```python
df_closing.plot()
```

```python
df_closing.iplot(kind='line',
                title='Apple, IBM, Netflix, Microsoft, and Oracle Close: 2010-8/2019')
```

```python
df_closing.to_csv('closing_stocks.csv')
```
