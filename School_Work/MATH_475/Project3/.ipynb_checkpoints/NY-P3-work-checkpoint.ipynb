## Imports

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Data 

```python
# Load data
asteroids = pd.read_csv("./Data/nasa.csv")
```

## Preprocessing

```python
# View info about columns
asteroids.info()
```

```python
# Describe columns
asteroids.describe()
```

```python
# check correlation between values
asteroids.corr()
```

```python
# Check for null values
sum(asteroids.isna().any())
```

```python
# Check object i
asteroids["Orbiting Body"].unique()
```

## Encode

```python
# Change T/F in target to 1/0
target_dummies = pd.get_dummies(data = asteroids.Hazardous)
```

```python
# Join target_dummies to dataframe
asteroids = asteroids.join(target_dummies)
```

```python
# drop hazardous column
asteroids = asteroids.drop(columns = ["Hazardous"], axis =1)
```

## Visualize

```python
sns.scatterplot(x= "Aphelion Dist", y = "Miles per hour", data = asteroids)
```
