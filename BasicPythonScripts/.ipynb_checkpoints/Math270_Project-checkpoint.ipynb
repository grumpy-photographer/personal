```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Read data
df = pd.read_csv('../Data/heart.csv')
df.head()
```

```python
# Clean data
df = df.drop(['cp','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target'], axis=1)

df = df.rename(columns={'age':'Age', 'sex':'Sex', 'trestbps':'Resting Blood Pressure (mm Hg)'})

df.head()
```
