# Part 2: Appending pandas Dataframes

### 2.1 Append the three Geo SAS Data Sets

- Rows: 1032  
- Columns:  53  

```python
# Append the three Geo SAS Data Sets
df_Appended_CA_OR = df_geo_ca.append(df_geo_or, ignore_index=True)
df_Appended_CA_OR_WA = df_Appended_CA_OR.append(df_geo_wa, ignore_index=True)

# Display number of rows and columns
print("Number of rows: ", df_Appended_CA_OR_WA.shape[0])
print("Number of columns: ", df_Appended_CA_OR_WA.shape[1])

df_Appended_CA_OR_WA.head()
```
