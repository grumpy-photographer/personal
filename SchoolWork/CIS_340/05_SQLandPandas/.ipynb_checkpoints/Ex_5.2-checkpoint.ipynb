# In Class  

### Build the Connection

```python
# Setup path and file name stuff

# Building Connection and get Cursor
current_directory = os.getcwd()

database_name = "Northwind_small.sqlite"

path_to_database = current_directory + "\Data\\" + database_name
print("path to database: " + path_to_database)

# Building Connection and get Cursor
conn = sqlite3.connect(path_to_database)
cursor = conn.cursor()
```

### Retrieve and Display All products which cost less than 20 Dollars
- Retrieve only these columns (with aliases, if shown) from the **Product** table  
  - Id as ProductID  
  - ProductName  
  - SupplierId as SupplierID  
  - CategoryId AS CategoryID  
  - QuantityPerUnit  
  - UnitPrice
  
- Order by price, lowest to highest

```python
sql_string =   """ select Id as ProductID, ProductName, SupplierId as SupplierID, CategoryId as CategoryID, QuantityPerUnit, UnitPrice from Product order by UnitPrice asc """

df_Products =pd.read_sql_query(sql_string, conn)

print("Number of Rows: ", df_Products.shape[0])
print("Number of Columns: ", df_Products.shape[1])
    
df_Products.head(10)
```

# Part 1  

### Retrieve and Display All Orders:  
1. From 2012  
2. Ordered by Order Date, oldest to newest  

Important Note: Because the name of the Order table is the same as an SQL Keyword, I needed to put left leaning apostrephes around Order in the from statement.

- Number of Rows:  152
- Number of Columns: 14 

```python
# Build the SQL String
sql_string = """  select * from 'Order' 
                    where OrderDate like '%2012%' 
                    Order by OrderDate asc
"""
df_Orders =pd.read_sql_query(sql_string, conn)

print("Number of Rows:  ", df_Orders.shape[0])
print("Number of Columns:  ", df_Orders.shape[1])

df_Orders.head(5)
```

### Retrieve and Display All Our North American Employees

```python
# Build the SQL String
sql_string = """ select * from Employee where Region = 'North America' """

df_Employees_NA =pd.read_sql_query(sql_string, conn)

print("Number of Rows:  ", df_Employees_NA.shape[0])
print("Number of Columns:  ", df_Employees_NA.shape[1])

df_Employees_NA.head(2)
```

```python
# Rename Id column to EmployeeId
df_Employees_NA = df_Employees_NA.rename({'Id': 'EmployeeId'}, axis = 'columns')

# Display top rows
df_Employees_NA.head(3)
```

# Part 2 
- Merge the **df_Orders** and **df_Employees_NA** dataframes to get additional Employee info columns in our new merged dataframe  

### Notes:  
1. You can create use the Template file to create a pickle and then display the pickle here, or you can add the merge code  
2. Either way, the last thing I want on this page is the verification code running successfully!


```python
# Do a Left Join 
df_left_join =  pd.merge(df_Orders, df_Employees_NA, on='EmployeeId', how='left')

df_left_join.head()
print("Number of Rows:  ", df_left_join.shape[0])
print("Number of Columns:  ", df_left_join.shape[1])
```

### Verify that the Left Join was correct

```python
# Verify that the Left Join was correct
# Verify number of columns
left_cols = df_Orders.shape[1]
right_cols = df_Employees_NA.shape[1]

expected_merge_cols = left_cols + right_cols - 1
actual_merge_cols = df_left_join.shape[1]


# Check number of rows
print("Expected Number of Rows:  ", df_Orders.shape[0])
print("Actual Number of Rows: ", df_left_join.shape[0])

if  df_left_join.shape[0] == df_Orders.shape[0]:
    print("Number of Rows in the Merged Dataframe is Correct!!!")
else:
    print("Number of Columns in the Merged Dataframe is NOT Correct!!!")
 
assert df_left_join.shape[0] == df_Orders.shape[0], "The Number of Rows in the Merge is NOT Correct!!"

# Check number of columns
print()
print("Expected Number of Columns:  ", expected_merge_cols)
print("Actual Number of Columns: ", actual_merge_cols)

if df_left_join.shape[1] == expected_merge_cols:
    print("Number of Columns in the Merged Dataframe is Correct!!!")
else:
    print("Number of Columns in the Merged Dataframe is NOT Correct!!!")
 

assert actual_merge_cols == expected_merge_cols, "The Number of Columns in the Merge is NOT Correct!!"
```

### Save the Left Join/Merge as a pickle  
- Name the pickle:  Ex_5.2_Part2.pkl

```python
# Save the df dataframe as a pickle file in the Data folder
df_left_join.to_pickle('Data/Ex_5.2_Part2.pkl')

# Read the new pickle file for a check
df_Test = pd.read_pickle('./Data/Ex_5.2_Part2.pkl')

# Display the top rows of the test dataframe
df_Test.head(3)
```
