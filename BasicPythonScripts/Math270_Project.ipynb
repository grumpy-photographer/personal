```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Read data
df = pd.read_csv('../Data/heart.csv')
df.head()
```

```python
# Clean data
df = df.drop(['cp','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target'], axis=1)

df = df.rename(columns={'age':'Age', 'sex':'Sex #', 'trestbps':'Resting Blood Pressure (mm Hg)'})

df.loc[df['Sex #'] == 1, 'Sex'] = 'M'
df.loc[df['Sex #'] == 0, 'Sex'] = 'F'

df.head()
```

```python
# 5 Number Summary (and then some)
df.describe()
```

```python
# Correlation
df.corr()
```

```python
# Covariance
df.cov()
```

```python
import matplotlib.pyplot as plt

x = df['Resting Blood Pressure (mm Hg)']
y = df['Age']
coef = np.polyfit(x, y,1)
poly1d_fn = np.poly1d(coef)

plt.plot(x, y, 'yo', x, poly1d_fn(x), '--k')
#df = sns.lineplot(df['Resting Blood Pressure (mm Hg)'], df['Age'], ci=95)
```
