# In Class  

### products_categories merge data  
- Data File: **w3schools_products_categories.pkl**  


```python
# Read the pickle:  w3schools_products_categories.pkl
file = './Data/Ex_4.4_InClass_Merge.pkl'

df_pc = pd.read_pickle(file)

#Display the number of rows and the top rows of the dataframe
print("Number of rows: ", df_pc.shape[0])
df_pc.head()
```

```python
# Create a new Calculated Column:  EstimatedTax
# EstimatedTax = Price * .10
df_pc['EstimatedTax'] = (df_pc.Price * .10)
# Display top rows
df_pc.head()
```

```python
# Create a new dataframe filled with only the product rows that are in the Condiments category
filter = df_pc['CategoryName'] == 'Condiments'
df_condiments = df_pc[filter]
print("Number of rows: ", df_condiments.shape[0])

# Display the number of rows and top rows in the dataframe
df_condiments.head()
```

```python
# Display the sum of all the Prices values in Condiments category
condiment_sum = df_condiments['Price'].sum()

print(f'The sum of all the prices in condiments is: ${condiment_sum:,.2f}')
```

```python
# Display the percentage of the value of our inventory that are condiments
total_sum = df_pc['Price'].sum()
percentage = (condiment_sum / total_sum) * 100
print(f'The percentage of the value of the inventory that are condiments is: {percentage:.1f}%')
```

### Shoe Store Data 

1. File name: **ShoeStore_Large.xlsx**  
2. Worksheet name:  **Large Data**  
3. Rename Price ($) and Size (US) to Price and Size   
4. Check data types for all the columns. Change as needed.



```python
# Read the Large Data worksheet data into a dataframe named df_shoes
file = 'Data//ShoeStore_Large.xlsx'
df_shoes = pd.read_excel(file, 'Large Data')

# Display the number of rows and top rows in the dataframe
print("Number of rows: ", df_shoes.shape[0])
df_shoes.head()
```

```python
# Drop the Returned and Gender columns
df_shoes.drop(['Returned', 'Gender'], axis='columns', inplace=True) 
```

```python
# Display the df_shoes columns
df_shoes.columns
```

```python
# Rename Price ($) to Price and Size (US) to Size
df_shoes.rename(columns = {'Price ($)':'Price'}, inplace=True)
df_shoes.rename(columns = {'Size (US)': 'Size'}, inplace=True)
# Display top rows
df_shoes.head()
```

```python
# Create the following new columns:  quarter_number, month_number,  weekday_number and weekday_name
df_shoes['quarter_number'] = df_shoes['Order Date'].dt.to_period("Q")
df_shoes['month_number'] = df_shoes['Order Date'].dt.month
df_shoes['weekday_number'] = df_shoes['Order Date'].dt.weekday
df_shoes['weekday_name'] = df_shoes['Order Date'].dt.weekday_name

df_shoes.head()
```
