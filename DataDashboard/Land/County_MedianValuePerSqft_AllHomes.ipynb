```python
#Imports
import pandas as pd
```

```python
#Load Land data
df_mvsf = pd.read_excel('./Data/County_MedianValuePerSqft_AllHomes.xlsx')

#Display table to ensure data loaded correctly
df_mvsf.head()
```

```python
#Filter data to NC
filter1 = df_mvsf['State'] == "NC"
df_mvsf_nc = df_mvsf[filter1]

#Check to ensure filter worked
df_mvsf_nc.head(5)
```

```python
#Change MunicipalCodeFIPS dtype to add leading 0's
df_mvsf_nc['MunicipalCodeFIPS'] = df_mvsf_nc['MunicipalCodeFIPS'].astype(str)
```

```python
#Add leading 0's and check to ensure they were added
df_mvsf_nc['MunicipalCodeFIPS'] = df_mvsf_nc['MunicipalCodeFIPS'].str.zfill(3)
df_mvsf_nc.head(5)
```

```python
#Save to csv file for export in Excel
df_mvsf_nc.to_csv('./Data/STG_ZLLW_County_MedianValuePerSqft_AllHomes.txt')
```
