```python
import pandas as pd
import seaborn as sns
```

```python
housing_df = pd.read_csv("california-housing.csv")
housing_df.columns
```

```python
sns_plot = sns.scatterplot(x = "longitude",y = "latitude",data = housing_df, hue = "population").set_title("Population by Block")
```
