```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
%matplotlib inline
```

```python
# Shapefile Contents #
# shp -> shapefile -> geometry
# dbf -> database -> extra info
# prj -> projection
# shx -> index
```

# Shapefile 

```python
# shp
df = gpd.read_file("C:/Users/natha/OneDrive/Desktop/GitHub/personal/Data/states.shp")
#df.head()
```

```python
df.plot()
```

```python
df.to_crs({'proj':'merc'}).plot()
```

```python
df.to_crs({
    'datum':'NAD83',
    'no_defs':True, 
    'proj':'aea',
    'lat_1':29.5,
    'lat_2':25.5,
    'lat_0':23,
    'lon_0':-96
}).plot()
```

```python
df.to_crs(epsg=5071).plot()
```
