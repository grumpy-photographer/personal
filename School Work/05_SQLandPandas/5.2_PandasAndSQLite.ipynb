```python
import pandas as pd
import sqlite3
import os 
```

# Building the Connection

```python
# Setup path and file name stuff
current_directory = os.getcwd()

database_name = "Northwind_small.sqlite"

path_to_database = current_directory + "\Data\\" + database_name
print("path to database: ", path_to_database)

# Building Connection and get Cursor
conn = sqlite3.connect(path_to_database)
cursor = conn.cursor()
```

# Listing all the Tables in an SQLite Database

```python
df_tables = pd.read_sql_query("select name from sqlite_master where type='table'", conn)
df_tables.sort_values('name', inplace=True)
df_tables
```

# Filling a Dataframe with a *SELECT* command

```python
# Build the SQL String
sql_string = """ 
              select * from employee   
            """
# Execute a query on the database and fill a dataframe with the result
df_employees = pd.read_sql_query(sql_string, conn)

# Display top rows
df_employees.head(2)
```

```python
# Show some info about the dataframe:
print("The number of rows in the dataframe are:  ", df_employees.shape[0])
print("The number of columns in the dataframe are:  ", df_employees.shape[1])
print()
print("Dataframe Columns:  ", df_employees.columns)
print()
print("Datatypes:")
print(df_employees.dtypes)
```

# Using a WHERE command to Subset the Dataframe  
- Display Product records with CategoryID 1, 2 or 3  
- Sort by id

```python
# Build the SQL String
sql_string = """ 
               select * from Product
                   where CategoryId in (1, 2, 3)
                   order by CategoryId
            """

df_Products =pd.read_sql_query(sql_string, conn)
df_Products.head(10)
```

```python
# Show some info about the dataframe: df_Products
print("The name of the dataframe is:  df_Products")
print()
print("The number of rows in the dataframe are:  ", df_Products.shape[0])
print("The number of columns in the dataframe are:  ", df_Products.shape[1])
print("Columns:  ", df_Products.columns)
print()
print("Datatypes:")
df_Products.dtypes
```
